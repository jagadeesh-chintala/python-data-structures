```python
def fun1(name,age=25):
    print("name:",name)
    print("age:",age)
fun1("jaggu",21)
```

```python
#functio with parameters
def add(a, b):
    return a + b

result = add(10, 20)
print(result)
```

```python
#default parameters
def greet(name="Guest"):
    print("Hello,", name)

greet()
greet("John")
```

```python
#keyword arguments
def student(name, age):
    print("Name:", name)
    print("Age:", age)

student(age=20, name="Emma")
```

```python
#variable length
def add(*numbers):
    print(sum(numbers))

add(1, 2)
add(1, 2, 3, 4)
```

```python
palindrome = lambda s: s == s[::-1]

text = "madam"

if palindrome(text):
    print("Palindrome")
else:
    print("Not Palindrome")
```
