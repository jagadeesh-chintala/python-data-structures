```python
#given a positive no 'n' greater than 2 create a numpy array of size (n*n) with all zeros and ones such that the ones make a shape of plus

a2 = np.zeros((n, n), dtype=int)

mid = n // 2
a2[mid, :] = 1     
a2[:, mid] = 1     

print(a2)
```
